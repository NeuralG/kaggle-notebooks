```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("WineQT.csv")
```

```python
quality_categories = sorted(df["quality"].unique())
quality_type = pd.CategoricalDtype(categories=quality_categories, ordered=True)
df["quality"] = df["quality"].astype(quality_type)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
sns.countplot(x=df["quality"])
```

```python
sns.displot(data=df, x="fixed acidity")
```

```python
sns.histplot(data=df[df["quality"] >= 6], x="fixed acidity")
```

```python
sns.kdeplot(data=df, x="fixed acidity", hue=df["quality"] >= 6, alpha=0.6)
```

```python
sns.kdeplot(data=df, x="volatile acidity", hue=df["quality"] >= 6, alpha=0.6)
```

Fixed aciditiy is not predictive but volatile acidity is slightly negative correlated with quality.


```python
sns.kdeplot(data=df, x="citric acid", hue=df["quality"] >= 6, alpha=0.6)
```

```python
df.groupby("quality")["citric acid"].describe()
```

There is positive correlation with citric acid and quality


```python
np.corrcoef(df["quality"], df["residual sugar"])
```

```python
sns.kdeplot(data=df, x="residual sugar", hue=df["quality"] <= 5)
```

Residual sugar and quality have no correlation.


```python
np.corrcoef(df["chlorides"], df["residual sugar"])
```

```python
df.groupby("quality")["chlorides"].describe()
```

When chloride decreases, quality is generally increases


```python
df.groupby("quality").mean()
```

```python
matrix = df.corr()
sns.heatmap(matrix)
```

```python
df["sulfur ratio"] = df["free sulfur dioxide"] / df["total sulfur dioxide"]
```

```python
df.groupby("quality")["sulfur ratio"].describe()
```

```python
sns.boxplot(data=df, x="sulfur ratio", y="quality")
```

Sulfur ratio is not predictive.


```python
df.groupby("quality")["free sulfur dioxide"].describe()
```

```python
df.groupby("quality")["total sulfur dioxide"].describe()
```

```python
sns.boxplot(data=df, x="total sulfur dioxide", y="quality")
```

Both of the sulfur features are not predictive.


```python
sns.boxplot(data=df, x="sulphates", y="quality")
```

```python
sns.boxplot(data=df, x="alcohol", y="quality")
```

Sulphates and alcohol are predictive but, for quality = 5, there are lots of outliers so we cant dedect from this alone.


All findings:

-   Sulphates and alcohol are predictive but, for quality = 5, there are lots of outliers so we cant dedect from this alone.
-   Both of the sulfur features are not predictive.
-   Sulfur ratio is not predictive.
-   When chloride decreases, quality is generally increases
-   Residual sugar and quality have no correlation.
-   There is positive correlation with citric acid and quality
-   Fixed aciditiy is not predictive but volatile acidity is slightly negative correlated with quality.
